#### Questao 1) - Separar a parte Par e a parte Ímpar de um Sinal

```python
def Ajustar_Sinal_Tempo_(Sinal, N, Novo_N):
  """
  Funcao para ajustar um sinal
  em relacao a um tempo (Novo_N).
  
  Entrada:
    - Vetor de pontos do sinal.
    - Vetor de pontos so sinal em relacao ao tempo.
    - Vetor ajustado de N.
  Saida:
    - Vetor de pontos do sinal em relacao ao tempo ajustado.
  """
  
  Novo_S = Np.zeros(len(Novo_N), dtype = int)   # Vetor de ajuste do sinal.

  # Loop para ajustar o sinal em relacao ao tempo.
  for i in range(len(Sinal)):

    Index = Np.where(Novo_N == N[i])  # Novo_N[?] = N[i].
    Novo_S[Index] = Sinal[i]          # Ajuste do sinal no tempo.

  return (Novo_S)
```

```python
def Espelhar_Sinal_(Sinal, N):
  """
  Funcao para espelhar um sinal.
  
  Entrada:
    - Vetor de pontos do sinal.
    - Vetor de pontos do sinal em relacao ao tempo.
  Saida:
    - Vetor ajustado de pontos do sinal.
    - Vetor ajustado de pontos do espelho do sinal.
    - Vetor ajustado de N.
  """

  Espelho_N     = Np.flip(N)      # Espelho de N.
  Espelho_Sinal = Np.flip(Sinal)  # Espelho do sinal.

  # Verifica se o espelho de N eh diferente de N.
  # (N nao eh simetrico em relacao a origem). 
  if(Np.array_equal(Espelho_N, N) != 1):

    Max    = max(abs(max(N)), abs(min(N)))  # Maior valor absoluto em N.
    Novo_N = Np.arange(-Max, Max+1)         # Vetor ajustado de N.
    Espelho_N *= -1                         # Vetor ajustado de -N.

    Sinal         = Ajustar_Sinal_Tempo_(Sinal, N, Novo_N)                   # Vetor ajustado do sinal no tempo.
    Espelho_Sinal = Ajustar_Sinal_Tempo_(Espelho_Sinal, Espelho_N, Novo_N)   # Vetor ajustado do espelho do sinal no tempo.
    N = Novo_N

  return (Sinal, Espelho_Sinal, N)
```

```python
def Separar_Sinal_(Sinal, Espelho_Sinal):
  """
  Funcao para separar a parte par
  e a parte impar de um sinal.

  Entrada:
    - Vetor de pontos do sinal.
    - Vetor de pontos do espelho do sinal.
  Saida:
    - Parte par e impar do sinal.
  """
  
  Sp = (Sinal + Espelho_Sinal)/2  # Parte par do sinal.
  Si = (Sinal - Espelho_Sinal)/2  # Parte impar do sinal.

  return (Sp, Si)
```

```python
def Plotar_Sinal_(Sinal, Sp, Si, N):
  """
  Funcao para plotar um sinal com a sua parte par e impar.

  Entrada:
    - Vetor de pontos do sinal.
    - Vetor de pontos da parte par do sinal.
    - Vetor de pontos da parte impar do sinal.
    - Vetor de pontos do sinal em relacao ao tempo.
  Saida:
    - None.
  """

  Fig, Tabs = Graph.subplots(2, 2, sharex = True)
  Graph.subplots_adjust(wspace = 0.5, hspace = 1)


  # Subplote do sinal.
  Tabs[0][0].stem(N, Sinal, use_line_collection = True)
  Tabs[0][0].set_title("Sinal")
  Tabs[0][0].set_ylabel("Amostras")
  Tabs[0][0].grid()

  # Subplote do sinal: par + impar.
  Soma = (Sp + Si)
  Tabs[0][1].stem(N, Soma, use_line_collection = True)
  Tabs[0][1].set_title("Sinal:  Par + Impar")
  Tabs[0][1].grid()

  # Subplote da parte par do sinal.
  Tabs[1][0].stem(N, Sp, use_line_collection = True)
  Tabs[1][0].set_title("Sinal:  Par")
  Tabs[1][0].set_ylabel("Amostras")
  Tabs[1][0].set_xlabel("n")
  Tabs[1][0].grid()

  # Subplote da parte impar do sinal.
  Tabs[1][1].stem(N, Si, use_line_collection = True)
  Tabs[1][1].set_title("Sinal:  Impar")
  Tabs[1][1].set_xlabel("n")
  Tabs[1][1].grid()

  Fig.tight_layout()
```

```python
# Letra A.
S = [ 2,  1, 0, 1, 2]
N = [-2, -1, 0, 1, 2]

S, Espelho_S, N = Espelhar_Sinal_(S, N)
Sp, Si = Separar_Sinal_(S, Espelho_S)
Plotar_Sinal_(S, Sp, Si, N)
```

#### Questao 2) - Gerar Sinal Degrau

```python
def Gerar_Sinal_Degrau_(N, K = 0):
  """
  Funcao para gerar um sinal degrau.

  Entrada:
    - Vetor de pontos em relacao ao tempo.
    - Indice para o deslocamento no tempo:
        * K < 0 ---> Direita.
        * K > 0 ---> Esquerda.
  Saida:
    - Sinal degrau deslocado no tempo.
  """

  N = N + K                               # Deslocamento de N no tempo.
  Degrau = Np.zeros(len(N), dtype = int)  # Vetor do sinal degrau.
  Degrau[N >= 0] = 1                      # Sinal degrau.

  return Degrau
```

#### Questao 3) - Gerar Sinal de uma Onda Quadrada

```python
def Gerar_Sinal_Onda_Quadrada_(Funcao):
  """
  Funcao para gerar o sinal de uma onda quadrada.

  Entrada:  Valor de uma funcao quadrada.
  Saida:    Sinal de uma onda quadrada.
  """
  return Sig.square(Funcao)
```

```python
def Sistema_H_(N):
  """
  Funcao para gerar um sinal de resposta a um
  vetor de pontos no tempo aplicado ao sistema H.

  Entrada:  Vetor de pontos em relacao ao tempo.
  Saida:    Sinal de resposta ao sistema H.
  """

  # Sistema H:
  #   H = 1, se 0 <= N <= 2.
  #   H = 0, caso contrario.
  H = Np.zeros(len(N), dtype = int)
  for i in range(len(N)):
    if(0 <= N[i] <= 2):
      H[i] = 1
  
  return H
```

```python
def Sistema_X_(N):
  """
  Funcao para gerar um sinal de resposta a um
  vetor de pontos no tempo aplicado ao sistema X.

  Entrada:  Vetor de pontos em relacao ao tempo.
  Saida:    Sinal de resposta ao sistema X.
  """

  # Sistema X:
  #   X = 0.5, se N = 0.
  #   X = 2.0, se N = 1.
  #   X = 0.0, caso contrario.
  X = Np.zeros(len(N))
  X[N == 0] = 0.5
  X[N == 1] = 2

  return X
```

```python
# Teste da funcao.
N = Np.arange(-4, 6)
H = Sistema_H_(N)
X = Sistema_X_(N)
Y = Np.convolve(H, X, 'same')

Fig, Tabs = Graph.subplots(3, 1)
Graph.subplots_adjust(wspace = 0.5, hspace = 1)


# Subplote do sistema H[n].
Tabs[0].stem(N, H, use_line_collection = True)
Tabs[0].set_title("Sistema H[n]")
Tabs[0].set_ylim(-0.5, 1.5)
Tabs[0].grid()

# Subplote do sistema X[n].
Tabs[1].stem(N, X, use_line_collection = True)
Tabs[1].set_title("Sistema X[n]")
Tabs[1].set_ylabel("Amostra")
Tabs[1].set_ylim(-0.5, 2.5)
Tabs[1].grid()

# Subplote da convolucao Y[n] = H[n] * X[n].
Tabs[2].stem(N, Y, use_line_collection = True)
Tabs[2].set_title("Y[n] = H[n] * X[n]")
Tabs[2].set_xlabel("n")
Tabs[2].set_ylim(-0.5, 3.5)
Tabs[2].grid()
```

```python
def Gerar_Impulso_Unitario_(N, K = 0):
  """
  Funcao para gerar um impulso unitario.

  Entrada:
    - Vetor de pontos em relacao ao tempo.
    - Indice para o deslocamento no tempo:
        * K < 0 ---> Direita.
        * K > 0 ---> Esquerda.
  Saida:
    - Impulso unitario deslocado no tempo.
  """

  N = N + K                                 # Deslocamento de N no tempo.
  Impulso = Np.zeros(len(N), dtype = int)   # Vetor do impulso.
  Impulso[N == 0] = 1                       # Impulso.

  return Impulso
```
